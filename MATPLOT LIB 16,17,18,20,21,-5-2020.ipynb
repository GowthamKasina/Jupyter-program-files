# PIE CHART

```python
# PIE CHART
# RENT = 3000

expenses = [1400,600,300,410,290]
expenses_labels = ["HOME RENT","FOOD","PHONE","BIKE","DONATIONS"]
plt.pie(expenses,labels = expenses_labels)
```

```python
expenses = [1400,600,300,410,290]
expenses_labels = ["HOME RENT","FOOD","PHONE","BIKE","DONATIONS"]
plt.pie(expenses,labels = expenses_labels)
plt.axis("equal")
plt.show()
```

```python
expenses = [1400,600,300,410,290]
expenses_labels = ["HOME RENT","FOOD","PHONE","BIKE","DONATIONS"]
plt.pie(expenses,labels = expenses_labels, shadow = True)
plt.axis("equal")
plt.show()
```

```python
# Whenever you are showing calling PIE with n labels, it will not in the form of circle, it will be in eclipse format
# Show the percentage of every pie, increase the chart size.

expenses = [1400,600,300,410,290]
expenses_labels = ["HOME RENT","FOOD","PHONE","BIKE","DONATIONS"]
plt.pie(expenses,labels = expenses_labels, shadow = True,autopct = '%1.2f%%') #1.2f means two floting points
plt.axis("equal")
plt.show()
```

```python
expenses = [1400,600,300,410,290]
expenses_labels = ["HOME RENT","FOOD","PHONE","BIKE","DONATIONS"]
plt.pie(expenses,labels = expenses_labels,autopct = '%1.1f%%') #1.1f means one floting point
plt.axis("equal")
plt.show()
```

```python
expenses = [1400,600,300,410,290]
expenses_labels = ["HOME RENT","FOOD","PHONE","BIKE","DONATIONS"]
plt.pie(expenses,labels = expenses_labels,autopct = '%1.1f%%', shadow = True, radius = 2.5) 
plt.show()
```

```python
expenses = [1400,600,300,410,290]
expenses_labels = ["HOME RENT","FOOD","PHONE","BIKE","DONATIONS"]
plt.pie(expenses,labels = expenses_labels,autopct = '%1.1f%%',radius = 2.5,explode = (0,0.1,0,0,0.1))
plt.show()
```

# HOW TO SAVE PIE CHART

```python
# HOW TO SAVE PIE CHART
expenses = [1600,600,300,410,290]
plt.pie(expenses,labels = expenses_labels,radius = 2,autopct = "%0.1f%%",explode = [0,0.1,0,0,0])
plt.axis("equal")
```

```python
expenses = [1600,600,300,410,290]
expenses_labels = ["Home_rent","Food","InternetBill","Bike","Donations"]
plt.pie(expenses,labels = expenses_labels,radius = 2,autopct = "%0.1f%%",explode = [0,0.1,0,0,0])
plt.axis("equal")
plt.savefig("piechart.jpg")
plt.show()
```

```python
expenses = [1600,600,300,410,290]
expenses_labels = ["Home_rent","Food","InternetBill","Bike","Donations"]
plt.pie(expenses,labels = expenses_labels,radius = 2,autopct = "%0.1f%%",explode = [0,0.1,0,0,0])
plt.axis("equal")
plt.savefig("piechart.jpg",transparent = True)
plt.show()
```

```python
expenses = [1600,600,300,410,290]
expenses_labels = ["Home_rent","Food","InternetBill","Bike","Donations"]
plt.pie(expenses,labels = expenses_labels,radius = 2,autopct = "%0.1f%%",explode = [0,0.1,0,0,0])
plt.axis("equal")
plt.savefig("piechart.jpg",bbox_inches = "tight",transparent = True)
plt.show()
```
